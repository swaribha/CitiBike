### Combining the Ridership Data for 2019

```python
file_riderdata=glob.glob('RidershipdataLaunch-2019q2/*/*.csv')
file_riderdata
```

```python
df_riderData = pd.concat([pd.read_csv(f) for f in glob.glob('RidershipdataLaunch-2019q2/*/*.csv')], ignore_index=True)
df_riderData
```

```python
# print(df_riderData.columns)
print(df_riderData.dtypes)
```

### Column name changed Over the years so merging data and deleting one of the column

```python
df_riderData['24-Hour Passes Purchased (midnight to 11:59 pm)']=df_riderData['24-Hour Passes Purchased (midnight to 11:59 pm)'].combine_first(df_riderData['24-Hour Passes Purchased (midnight to 11:59 pm)\t7-Day Passes Purchased (midnight to 11:59 pm)'])
df_riderData
```

```python
df_riderData['Cumulative trips (since launch)']=df_riderData['Cumulative trips (since launch)'].combine_first\
                                        (df_riderData['Cumulative trips (since launch):'])
# df_riderData
```

```python
df_riderData['Cumulative trips (since launch)']=df_riderData['Cumulative trips (since launch)'].combine_first\
                                        (df_riderData['Cumulative trips (since launch):'])
```

```python
df_riderData['Miles traveled to date']=df_riderData['Miles traveled to date'].combine_first(df_riderData['Miles traveled to date:'])    
```

```python
df_riderData['Total Annual Members']=df_riderData['Total Annual Members'].combine_first(df_riderData['Total Annual Members (All Time)'])
 
```

```python
df_riderData.drop(columns=["Column 1",\
                           "24-Hour Passes Purchased (midnight to 11:59 pm)\t7-Day Passes Purchased (midnight to 11:59 pm)",\
                           "Cumulative trips (since launch):",\
                          "Miles traveled to date:",\
                          "Total Annual Members (All Time)"],inplace=True)
```

```python
df_riderData.drop(columns=["Total Annual Memberships Sold"],inplace=True)
```
